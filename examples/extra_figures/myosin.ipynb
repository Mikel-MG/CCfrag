```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

# EEA1

```python
list_specifications = ['2_20_10',
                      '2_30_15',
                      '2_50_25',
                      '2_70_35',]
```

```python
df_ccfrag_scores = pd.read_csv('../GitHub_repo/CCfrag/examples/EEA1/CCfrag_constructs_EEA1/EEA1_HS/DATA_CCfrag.csv', index_col=0)
```

```python
list_scores = []

target_df = df_ccfrag_scores[(df_ccfrag_scores['residue_number'] > 200) & (df_ccfrag_scores['residue_number'] < 1200)]
target_df = df_ccfrag_scores

for specification in list_specifications:
    scores = target_df[f'{specification}_plddt']
    list_scores.append(scores)
    plt.hist(scores, alpha=0.25, bins=25)
```

```python
plt.plot(list_scores, color='black', alpha=0.075)
None
```

```python
colors = [f'C{i}' for i in range(4)]
median_color='black'

outlier_color='gray'
outlier_alpha=1
```

```python
fig, ax = plt.subplots(figsize=(6,4), dpi=200)
box = ax.boxplot(np.array(list_scores).T, patch_artist=True,
                flierprops=dict(marker='o', markeredgecolor=outlier_color, alpha=outlier_alpha, markersize=6,markeredgewidth=0.5))

# change the color of the boxes to match the other plots
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.55)

# Change the color of the median lines
for median in box['medians']:
    median.set_color(median_color)
    median.set_linewidth(1)  # Optionally, adjust the line width

ax.set_ylabel('pLDDT')
ax.set_xlabel('CCfrag specification')
ax.set_xticks(ticks=range(1,5), labels=list_specifications)
plt.tight_layout()
plt.show()
```

```python
from mikel_tools.PDB.io import extract_bfactor
```

```python
eea1_plddt = extract_bfactor('../main_examples/1_EEA1/__full/unrelaxed_model_1_multimer_v3_pred_0.pdb')
```

```python
plt.plot(eea1_plddt)
```

```python
list_scores = [i.to_numpy() for i in list_scores]
```

```python
all_colors = ['lightgray'] + colors
```

```python
fig, ax = plt.subplots(figsize=(6,4), dpi=200)

widths = 0.5

all_data = [eea1_plddt] + list_scores

#all_data = list_scores + [list_scores[0]]

box = ax.boxplot(all_data, patch_artist=True,
                flierprops=dict(marker='o', markeredgecolor=outlier_color, alpha=outlier_alpha, markersize=6,markeredgewidth=0.5), positions=range(0,5), widths=widths)

# change the color of the boxes to match the other plots
for patch, color in zip(box['boxes'], all_colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.55)

# Change the color of the median lines
for median in box['medians']:
    median.set_color(median_color)
    median.set_linewidth(1)  # Optionally, adjust the line width

ax.set_ylabel('pLDDT')
ax.set_xlabel('CCfrag specification')
ax.set_xticks(ticks=range(0,5), labels=['Full-length'] + list_specifications)
plt.tight_layout()
plt.show()
```
